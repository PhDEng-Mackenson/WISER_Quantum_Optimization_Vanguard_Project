# 1000 iterations 

```python
# ====== Imports ======
import json
import numpy as np
import matplotlib.pyplot as plt

from collections import defaultdict
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit.library import RealAmplitudes
from qiskit.primitives import Estimator, Sampler  
from qiskit_algorithms.utils import algorithm_globals
from qiskit_algorithms.minimum_eigensolvers import VQE
from qiskit_algorithms.optimizers import SPSA


# ====== User inputs ======
json_path = "bond_data1.json" 
MAX_BONDS = 19                  # target number of bonds (N)
SELECTED_BONDS = None         

algorithm_globals.random_seed = 42
np.random.seed(42)

# ====== Load & prepare data ======
with open(json_path, "r") as f:
    data = json.load(f)

# Optional overrides
data['cash_min'] = data.get('cash_min', 0.0)
data['cash_max'] = data.get('cash_max', 0.0)

all_bonds = data["bonds"]
sector_data = data["sector_data"]
cash_min = float(data["cash_min"])
cash_max = float(data["cash_max"])
cash_mid = 0.5*(cash_min + cash_max)

# Choose bonds
if SELECTED_BONDS:
    bonds = [b for b in all_bonds if b["id"] in SELECTED_BONDS]
elif MAX_BONDS:
    bonds = all_bonds[:MAX_BONDS]
else:
    bonds = all_bonds

print(f"Using {len(bonds)} bonds in the model.")

# Fixed trade quantity x_c and contributions _cj x_c
for b in bonds:
    basket_inv = b.get("basket_inventory", b["max"])
    delta_c    = b.get("delta_c", 1.0)
    eff_max    = min(b["max"], basket_inv)
    b["fixed_qty"]   = (b["min"] + eff_max) / (2.0*delta_c)
    b["pmv_contrib"] = b["pmv"] * b["fixed_qty"]
    b["dxs_contrib"] = b["dxs"] * b["fixed_qty"]

BENCHMARK_VALUE = sum(b["pmv_contrib"] for b in bonds) or 1.0 

# Group bonds by sector
sector_bonds = defaultdict(list)
for b in bonds:
    sector_bonds[b["sector"]].append(b)

num_qubits = len(bonds)

# ====== Utilities to build Ising penalties ======
def z_label(idxs, n):
    """Return a Pauli Z string 'ZZI...'. idxs are 0-based qubit indices."""
    s = ['I']*n
    for i in idxs:
        s[i] = 'Z'
    return ''.join(s)

def equality_penalty_ising(coeffs, target, weight, n):
    """
    Build *(_i c_i y_i - t)^2 as an Ising Hamiltonian using y_i = (1 - Z_i)/2.
    Returns a SparsePauliOp on n qubits.
    """
    if not coeffs:
        return SparsePauliOp.from_list([(z_label([], n), weight * (target**2))])

    cvals = np.array(list(coeffs.values()), dtype=float)
    idxs  = list(coeffs.keys())

    sum_c = float(cvals.sum())
    A     = 0.5*sum_c - float(target)

    # Constant: *(A^2 + 1/4  c_i^2)
    const = weight * (A*A + 0.25*float((cvals**2).sum()))
    H = SparsePauliOp.from_list([(z_label([], n), const)])

    # Linear terms: *(-A * c_i) Z_i
    for i, ci in zip(idxs, cvals):
        H += SparsePauliOp.from_list([(z_label([i], n), weight * (-A*ci))])

    # Pair terms: *(1/4 c_i c_j) Z_i Z_j  for i<j
    for a in range(len(idxs)):
        i, ci = idxs[a], cvals[a]
        for b in range(a+1, len(idxs)):
            j, cj = idxs[b], cvals[b]
            H += SparsePauliOp.from_list([(z_label([i,j], n), weight * (0.25*ci*cj))])

    return H

def constant_shift(op: SparsePauliOp) -> float:
    """Sum coefficients of identity terms (could be more than one after algebra)."""
    s = 0.0
    for lbl, coeff in zip(op.paulis.to_labels(), op.coeffs):
        if all(ch == 'I' for ch in lbl):
            s += float(np.real(coeff))
    return s


# ====== Weights======
WEIGHTS = {
    "pmv_objective":          36.0,
    "dxs_objective":          36.0,
    "max_bonds":            1000.0, 
    "cash_constraint":        50.0,
    "sector_pmv_constraint": 100.0,
    "sector_dxs_constraint": 100.0,
    "binary_constraint":     100.0,
}

# ====== Build Hamiltonian ======
H = SparsePauliOp.from_list([(z_label([], num_qubits), 0.0)])

# 1) Max bonds ( y_i  MAX_BONDS)
coeffs = {i: 1.0 for i in range(num_qubits)}
H += equality_penalty_ising(coeffs, MAX_BONDS, WEIGHTS["max_bonds"], num_qubits)

# 2) Residual cash ( pmv_contrib  mid) 
coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds)}
H += equality_penalty_ising(coeffs, cash_mid/BENCHMARK_VALUE, WEIGHTS["cash_constraint"], num_qubits)

# 3) Sector PMV band via midpoint (per sector)
for s, sdata in sector_data.items():
    coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    pmn, pmx = sdata.get("pmv_min"), sdata.get("pmv_max")
    if pmn is not None and pmx is not None:
        mid = 0.5*(pmn + pmx)/BENCHMARK_VALUE
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["sector_pmv_constraint"], num_qubits)

# 4) Sector DXS band via midpoint (per sector)
for s, sdata in sector_data.items():
    coeffs = {i: b["dxs_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    dmn, dmx = sdata.get("dxs_min"), sdata.get("dxs_max")
    if dmn is not None and dmx is not None:
        mid = 0.5*(dmn + dmx)/BENCHMARK_VALUE
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["sector_dxs_constraint"], num_qubits)

# 5) Binary count per sector (min/max via midpoint)
for s, sdata in sector_data.items():
    if "min_bonds" in sdata and "max_bonds" in sdata:
        coeffs = {i: 1.0 for i, b in enumerate(bonds) if b["sector"] == s}
        mid = 0.5*(sdata["min_bonds"] + sdata["max_bonds"])
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["binary_constraint"], num_qubits)

# === Objective: match sector targets ===
for s, sdata in sector_data.items():
    coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    H += equality_penalty_ising(coeffs, sdata["target"]/BENCHMARK_VALUE, WEIGHTS["pmv_objective"], num_qubits)

for s, sdata in sector_data.items():
    if "dxs_target" in sdata:
        coeffs = {i: b["dxs_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
        H += equality_penalty_ising(coeffs, sdata["dxs_target"]/BENCHMARK_VALUE, WEIGHTS["dxs_objective"], num_qubits)

H = H.simplify()
const_off = constant_shift(H)

print(f"\nHamiltonian built on {num_qubits} qubits.")
print(f"Constant shift (energy offset): {const_off:.6f}")

# ====== VQE setup======
ansatz = RealAmplitudes(num_qubits, entanglement="linear", reps=5)
optimizer = SPSA(maxiter=1000)
estimator = Estimator()

energy_hist = []

def callback(eval_count, params, value, std):
    energy_hist.append(float(np.real(value)))
    if eval_count % 10 == 0:
        print(f"Iter {eval_count:3d} | energy {value:.6f}")

initial_point = np.random.rand(ansatz.num_parameters)

vqe = VQE(estimator, ansatz, optimizer, initial_point=initial_point, callback=callback)
res = vqe.compute_minimum_eigenvalue(H)

print("\n=== VQE Result ===")
print(f"Optimal energy (incl. const): {res.eigenvalue.real:.6f}")
print(f"Optimal params length: {len(res.optimal_point)}")

# ====== Plots energy vs iteration ======
loss_hist = [e - const_off for e in energy_hist]

# Energy vs iteration
fig1, ax1 = plt.subplots(figsize=(8, 5))
ax1.plot(range(1, len(energy_hist) + 1), energy_hist, linewidth=2)
ax1.set_xlabel("Iteration")
ax1.set_ylabel("Energy")
ax1.set_title("VQE Energy vs. Iteration")
ax1.grid(True)
fig1.tight_layout()
plt.show()                           
fig1.savefig("energy_vs_iteration.png", dpi=180)

# Loss vs iteration
fig2, ax2 = plt.subplots(figsize=(8, 5))
ax2.plot(range(1, len(loss_hist) + 1), loss_hist, linewidth=2)
ax2.set_xlabel("Iteration")
ax2.set_ylabel("Loss (Energy - constant)")
ax2.set_title("VQE Loss vs. Iteration")
ax2.grid(True)
fig2.tight_layout()
plt.show()                             
fig2.savefig("loss_vs_iteration.png", dpi=180)

print("Saved: energy_vs_iteration.png, loss_vs_iteration.png")

# ====== Sample the optimized state and print top portfolios ======
sampler = Sampler()            
opt_circuit = ansatz.assign_parameters(res.optimal_point)

# Add classical bits + measurements
measured = opt_circuit.copy()
measured.measure_all()

shots = 10000
qres = sampler.run([measured], shots=shots).result()
qd = qres.quasi_dists[0]

num_bits = num_qubits
# Convert to probabilities dict
probs = (qd.binary_probabilities(num_bits=num_bits)
         if hasattr(qd, "binary_probabilities")
         else {(format(k, f"0{num_bits}b") if isinstance(k, int) else k): float(v)
               for k, v in dict(qd).items()})
counts = {b: int(round(p * shots)) for b, p in probs.items()}

def selected_indices_from_bitstring(bits):
    bits_be = bits[::-1] 
    return [i for i, bit in enumerate(bits_be) if bit == "1"]

top = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTop candidate portfolios:")
for rank, (bits, cnt) in enumerate(top, start=1):
    sel = selected_indices_from_bitstring(bits)
    pmv_by_sector = {s: 0.0 for s in sector_data}
    dxs_by_sector = {s: 0.0 for s in sector_data}
    total_pmv = 0.0
    count_by_sector = {s: 0 for s in sector_data}

    for i in sel:
        b = bonds[i]
        s = b["sector"]
        pmv_by_sector[s] += b["pmv_contrib"]
        dxs_by_sector[s] += b["dxs_contrib"]
        count_by_sector[s] += 1
        total_pmv += b["pmv_contrib"]

    print(f"\n#{rank}  freq={cnt/shots:.2%}  selected={len(sel)}  bitstring={bits}")
    print(f"   Total PMV: {total_pmv:,.2f}  (cash in [{cash_min:,.2f}, {cash_max:,.2f}])")

    for s, sd in sector_data.items():
        tgt = sd.get("target", 0.0)
        pmv = pmv_by_sector[s]
        dxs = dxs_by_sector[s]
        cnt_s = count_by_sector[s]
        print(f"   {s}: PMV={pmv:,.2f} (target {tgt:,.2f}) "
              f"| DXS={dxs:,.2f} "
              f"| count={cnt_s}")

# --- Final bond selection
best_bits = max(counts.items(), key=lambda kv: kv[1])[0]
best_sel = selected_indices_from_bitstring(best_bits)

print("\nBond Selection:")
for i, b in enumerate(bonds):
    print(f"  {b['id']}: {1 if i in best_sel else 0}")
```

```python
# ====== Imports ======
import json
import numpy as np
import matplotlib.pyplot as plt

from collections import defaultdict
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit.library import RealAmplitudes
from qiskit.primitives import Estimator, Sampler  
from qiskit_algorithms.utils import algorithm_globals
from qiskit_algorithms.minimum_eigensolvers import VQE
from qiskit_algorithms.optimizers import SPSA


# ====== User inputs ======
json_path = "bond_data1.json" 
MAX_BONDS = 19                  # target number of bonds (N)
SELECTED_BONDS = None         

algorithm_globals.random_seed = 42
np.random.seed(42)

# ====== Load & prepare data ======
with open(json_path, "r") as f:
    data = json.load(f)

# Optional overrides
data['cash_min'] = data.get('cash_min', 0.0)
data['cash_max'] = data.get('cash_max', 0.0)

all_bonds = data["bonds"]
sector_data = data["sector_data"]
cash_min = float(data["cash_min"])
cash_max = float(data["cash_max"])
cash_mid = 0.5*(cash_min + cash_max)

# Choose bonds
if SELECTED_BONDS:
    bonds = [b for b in all_bonds if b["id"] in SELECTED_BONDS]
elif MAX_BONDS:
    bonds = all_bonds[:MAX_BONDS]
else:
    bonds = all_bonds

print(f"Using {len(bonds)} bonds in the model.")

# Fixed trade quantity x_c and contributions _cj x_c
for b in bonds:
    basket_inv = b.get("basket_inventory", b["max"])
    delta_c    = b.get("delta_c", 1.0)
    eff_max    = min(b["max"], basket_inv)
    b["fixed_qty"]   = (b["min"] + eff_max) / (2.0*delta_c)
    b["pmv_contrib"] = b["pmv"] * b["fixed_qty"]
    b["dxs_contrib"] = b["dxs"] * b["fixed_qty"]

BENCHMARK_VALUE = sum(b["pmv_contrib"] for b in bonds) or 1.0 

# Group bonds by sector
sector_bonds = defaultdict(list)
for b in bonds:
    sector_bonds[b["sector"]].append(b)

num_qubits = len(bonds)

# ====== Utilities to build Ising penalties ======
def z_label(idxs, n):
    """Return a Pauli Z string 'ZZI...'. idxs are 0-based qubit indices."""
    s = ['I']*n
    for i in idxs:
        s[i] = 'Z'
    return ''.join(s)

def equality_penalty_ising(coeffs, target, weight, n):
    """
    Build *(_i c_i y_i - t)^2 as an Ising Hamiltonian using y_i = (1 - Z_i)/2.
    Returns a SparsePauliOp on n qubits.
    """
    if not coeffs:
        return SparsePauliOp.from_list([(z_label([], n), weight * (target**2))])

    cvals = np.array(list(coeffs.values()), dtype=float)
    idxs  = list(coeffs.keys())

    sum_c = float(cvals.sum())
    A     = 0.5*sum_c - float(target)

    # Constant: *(A^2 + 1/4  c_i^2)
    const = weight * (A*A + 0.25*float((cvals**2).sum()))
    H = SparsePauliOp.from_list([(z_label([], n), const)])

    # Linear terms: *(-A * c_i) Z_i
    for i, ci in zip(idxs, cvals):
        H += SparsePauliOp.from_list([(z_label([i], n), weight * (-A*ci))])

    # Pair terms: *(1/4 c_i c_j) Z_i Z_j  for i<j
    for a in range(len(idxs)):
        i, ci = idxs[a], cvals[a]
        for b in range(a+1, len(idxs)):
            j, cj = idxs[b], cvals[b]
            H += SparsePauliOp.from_list([(z_label([i,j], n), weight * (0.25*ci*cj))])

    return H

def constant_shift(op: SparsePauliOp) -> float:
    """Sum coefficients of identity terms (could be more than one after algebra)."""
    s = 0.0
    for lbl, coeff in zip(op.paulis.to_labels(), op.coeffs):
        if all(ch == 'I' for ch in lbl):
            s += float(np.real(coeff))
    return s


# ====== Weights======
WEIGHTS = {
    "pmv_objective":          36.0,
    "dxs_objective":          36.0,
    "max_bonds":            1000.0, 
    "cash_constraint":        50.0,
    "sector_pmv_constraint": 100.0,
    "sector_dxs_constraint": 100.0,
    "binary_constraint":     100.0,
}

# ====== Build Hamiltonian ======
H = SparsePauliOp.from_list([(z_label([], num_qubits), 0.0)])

# 1) Max bonds ( y_i  MAX_BONDS)
coeffs = {i: 1.0 for i in range(num_qubits)}
H += equality_penalty_ising(coeffs, MAX_BONDS, WEIGHTS["max_bonds"], num_qubits)

# 2) Residual cash ( pmv_contrib  mid) 
coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds)}
H += equality_penalty_ising(coeffs, cash_mid/BENCHMARK_VALUE, WEIGHTS["cash_constraint"], num_qubits)

# 3) Sector PMV band via midpoint (per sector)
for s, sdata in sector_data.items():
    coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    pmn, pmx = sdata.get("pmv_min"), sdata.get("pmv_max")
    if pmn is not None and pmx is not None:
        mid = 0.5*(pmn + pmx)/BENCHMARK_VALUE
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["sector_pmv_constraint"], num_qubits)

# 4) Sector DXS band via midpoint (per sector)
for s, sdata in sector_data.items():
    coeffs = {i: b["dxs_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    dmn, dmx = sdata.get("dxs_min"), sdata.get("dxs_max")
    if dmn is not None and dmx is not None:
        mid = 0.5*(dmn + dmx)/BENCHMARK_VALUE
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["sector_dxs_constraint"], num_qubits)

# 5) Binary count per sector (min/max via midpoint)
for s, sdata in sector_data.items():
    if "min_bonds" in sdata and "max_bonds" in sdata:
        coeffs = {i: 1.0 for i, b in enumerate(bonds) if b["sector"] == s}
        mid = 0.5*(sdata["min_bonds"] + sdata["max_bonds"])
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["binary_constraint"], num_qubits)

# === Objective: match sector targets ===
for s, sdata in sector_data.items():
    coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    H += equality_penalty_ising(coeffs, sdata["target"]/BENCHMARK_VALUE, WEIGHTS["pmv_objective"], num_qubits)

for s, sdata in sector_data.items():
    if "dxs_target" in sdata:
        coeffs = {i: b["dxs_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
        H += equality_penalty_ising(coeffs, sdata["dxs_target"]/BENCHMARK_VALUE, WEIGHTS["dxs_objective"], num_qubits)

H = H.simplify()
const_off = constant_shift(H)

print(f"\nHamiltonian built on {num_qubits} qubits.")
print(f"Constant shift (energy offset): {const_off:.6f}")

# ====== VQE setup======
ansatz = RealAmplitudes(num_qubits, entanglement="linear", reps=5)
optimizer = SPSA(maxiter=100)
estimator = Estimator()

energy_hist = []

def callback(eval_count, params, value, std):
    energy_hist.append(float(np.real(value)))
    if eval_count % 10 == 0:
        print(f"Iter {eval_count:3d} | energy {value:.6f}")

initial_point = np.random.rand(ansatz.num_parameters)

vqe = VQE(estimator, ansatz, optimizer, initial_point=initial_point, callback=callback)
res = vqe.compute_minimum_eigenvalue(H)

print("\n=== VQE Result ===")
print(f"Optimal energy (incl. const): {res.eigenvalue.real:.6f}")
print(f"Optimal params length: {len(res.optimal_point)}")

# ====== Plots energy vs iteration ======
loss_hist = [e - const_off for e in energy_hist]

# Energy vs iteration
fig1, ax1 = plt.subplots(figsize=(8, 5))
ax1.plot(range(1, len(energy_hist) + 1), energy_hist, linewidth=2)
ax1.set_xlabel("Iteration")
ax1.set_ylabel("Energy")
ax1.set_title("VQE Energy vs. Iteration")
ax1.grid(True)
fig1.tight_layout()
plt.show()                           
fig1.savefig("energy_vs_iteration.png", dpi=180)

# Loss vs iteration
fig2, ax2 = plt.subplots(figsize=(8, 5))
ax2.plot(range(1, len(loss_hist) + 1), loss_hist, linewidth=2)
ax2.set_xlabel("Iteration")
ax2.set_ylabel("Loss (Energy - constant)")
ax2.set_title("VQE Loss vs. Iteration")
ax2.grid(True)
fig2.tight_layout()
plt.show()                             
fig2.savefig("loss_vs_iteration.png", dpi=180)

print("Saved: energy_vs_iteration.png, loss_vs_iteration.png")

# ====== Sample the optimized state and print top portfolios ======
sampler = Sampler()            
opt_circuit = ansatz.assign_parameters(res.optimal_point)

# Add classical bits + measurements
measured = opt_circuit.copy()
measured.measure_all()

shots = 10000
qres = sampler.run([measured], shots=shots).result()
qd = qres.quasi_dists[0]

num_bits = num_qubits
# Convert to probabilities dict
probs = (qd.binary_probabilities(num_bits=num_bits)
         if hasattr(qd, "binary_probabilities")
         else {(format(k, f"0{num_bits}b") if isinstance(k, int) else k): float(v)
               for k, v in dict(qd).items()})
counts = {b: int(round(p * shots)) for b, p in probs.items()}

def selected_indices_from_bitstring(bits):
    bits_be = bits[::-1] 
    return [i for i, bit in enumerate(bits_be) if bit == "1"]

top = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTop candidate portfolios:")
for rank, (bits, cnt) in enumerate(top, start=1):
    sel = selected_indices_from_bitstring(bits)
    pmv_by_sector = {s: 0.0 for s in sector_data}
    dxs_by_sector = {s: 0.0 for s in sector_data}
    total_pmv = 0.0
    count_by_sector = {s: 0 for s in sector_data}

    for i in sel:
        b = bonds[i]
        s = b["sector"]
        pmv_by_sector[s] += b["pmv_contrib"]
        dxs_by_sector[s] += b["dxs_contrib"]
        count_by_sector[s] += 1
        total_pmv += b["pmv_contrib"]

    print(f"\n#{rank}  freq={cnt/shots:.2%}  selected={len(sel)}  bitstring={bits}")
    print(f"   Total PMV: {total_pmv:,.2f}  (cash in [{cash_min:,.2f}, {cash_max:,.2f}])")

    for s, sd in sector_data.items():
        tgt = sd.get("target", 0.0)
        pmv = pmv_by_sector[s]
        dxs = dxs_by_sector[s]
        cnt_s = count_by_sector[s]
        print(f"   {s}: PMV={pmv:,.2f} (target {tgt:,.2f}) "
              f"| DXS={dxs:,.2f} "
              f"| count={cnt_s}")

# --- Final bond selection
best_bits = max(counts.items(), key=lambda kv: kv[1])[0]
best_sel = selected_indices_from_bitstring(best_bits)

print("\nBond Selection:")
for i, b in enumerate(bonds):
    print(f"  {b['id']}: {1 if i in best_sel else 0}")
```

```python
# ====== Imports ======
import json
import numpy as np
import matplotlib.pyplot as plt

from collections import defaultdict
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit.library import RealAmplitudes
from qiskit.primitives import Estimator, Sampler  
from qiskit_algorithms.utils import algorithm_globals
from qiskit_algorithms.minimum_eigensolvers import VQE
from qiskit_algorithms.optimizers import SPSA


# ====== User inputs ======
json_path = "bond_data1.json" 
MAX_BONDS = 19                  # target number of bonds (N)
SELECTED_BONDS = None         

algorithm_globals.random_seed = 42
np.random.seed(42)

# ====== Load & prepare data ======
with open(json_path, "r") as f:
    data = json.load(f)

# Optional overrides
data['cash_min'] = data.get('cash_min', 0.0)
data['cash_max'] = data.get('cash_max', 0.0)

all_bonds = data["bonds"]
sector_data = data["sector_data"]
cash_min = float(data["cash_min"])
cash_max = float(data["cash_max"])
cash_mid = 0.5*(cash_min + cash_max)

# Choose bonds
if SELECTED_BONDS:
    bonds = [b for b in all_bonds if b["id"] in SELECTED_BONDS]
elif MAX_BONDS:
    bonds = all_bonds[:MAX_BONDS]
else:
    bonds = all_bonds

print(f"Using {len(bonds)} bonds in the model.")

# Fixed trade quantity x_c and contributions _cj x_c
for b in bonds:
    basket_inv = b.get("basket_inventory", b["max"])
    delta_c    = b.get("delta_c", 1.0)
    eff_max    = min(b["max"], basket_inv)
    b["fixed_qty"]   = (b["min"] + eff_max) / (2.0*delta_c)
    b["pmv_contrib"] = b["pmv"] * b["fixed_qty"]
    b["dxs_contrib"] = b["dxs"] * b["fixed_qty"]

BENCHMARK_VALUE = sum(b["pmv_contrib"] for b in bonds) or 1.0 

# Group bonds by sector
sector_bonds = defaultdict(list)
for b in bonds:
    sector_bonds[b["sector"]].append(b)

num_qubits = len(bonds)

# ====== Utilities to build Ising penalties ======
def z_label(idxs, n):
    """Return a Pauli Z string 'ZZI...'. idxs are 0-based qubit indices."""
    s = ['I']*n
    for i in idxs:
        s[i] = 'Z'
    return ''.join(s)

def equality_penalty_ising(coeffs, target, weight, n):
    """
    Build *(_i c_i y_i - t)^2 as an Ising Hamiltonian using y_i = (1 - Z_i)/2.
    Returns a SparsePauliOp on n qubits.
    """
    if not coeffs:
        return SparsePauliOp.from_list([(z_label([], n), weight * (target**2))])

    cvals = np.array(list(coeffs.values()), dtype=float)
    idxs  = list(coeffs.keys())

    sum_c = float(cvals.sum())
    A     = 0.5*sum_c - float(target)

    # Constant: *(A^2 + 1/4  c_i^2)
    const = weight * (A*A + 0.25*float((cvals**2).sum()))
    H = SparsePauliOp.from_list([(z_label([], n), const)])

    # Linear terms: *(-A * c_i) Z_i
    for i, ci in zip(idxs, cvals):
        H += SparsePauliOp.from_list([(z_label([i], n), weight * (-A*ci))])

    # Pair terms: *(1/4 c_i c_j) Z_i Z_j  for i<j
    for a in range(len(idxs)):
        i, ci = idxs[a], cvals[a]
        for b in range(a+1, len(idxs)):
            j, cj = idxs[b], cvals[b]
            H += SparsePauliOp.from_list([(z_label([i,j], n), weight * (0.25*ci*cj))])

    return H

def constant_shift(op: SparsePauliOp) -> float:
    """Sum coefficients of identity terms (could be more than one after algebra)."""
    s = 0.0
    for lbl, coeff in zip(op.paulis.to_labels(), op.coeffs):
        if all(ch == 'I' for ch in lbl):
            s += float(np.real(coeff))
    return s


# ====== Weights======
WEIGHTS = {
    "pmv_objective":          36.0,
    "dxs_objective":          36.0,
    "max_bonds":            1000.0, 
    "cash_constraint":        50.0,
    "sector_pmv_constraint": 100.0,
    "sector_dxs_constraint": 100.0,
    "binary_constraint":     100.0,
}

# ====== Build Hamiltonian ======
H = SparsePauliOp.from_list([(z_label([], num_qubits), 0.0)])

# 1) Max bonds ( y_i  MAX_BONDS)
coeffs = {i: 1.0 for i in range(num_qubits)}
H += equality_penalty_ising(coeffs, MAX_BONDS, WEIGHTS["max_bonds"], num_qubits)

# 2) Residual cash ( pmv_contrib  mid) 
coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds)}
H += equality_penalty_ising(coeffs, cash_mid/BENCHMARK_VALUE, WEIGHTS["cash_constraint"], num_qubits)

# 3) Sector PMV band via midpoint (per sector)
for s, sdata in sector_data.items():
    coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    pmn, pmx = sdata.get("pmv_min"), sdata.get("pmv_max")
    if pmn is not None and pmx is not None:
        mid = 0.5*(pmn + pmx)/BENCHMARK_VALUE
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["sector_pmv_constraint"], num_qubits)

# 4) Sector DXS band via midpoint (per sector)
for s, sdata in sector_data.items():
    coeffs = {i: b["dxs_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    dmn, dmx = sdata.get("dxs_min"), sdata.get("dxs_max")
    if dmn is not None and dmx is not None:
        mid = 0.5*(dmn + dmx)/BENCHMARK_VALUE
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["sector_dxs_constraint"], num_qubits)

# 5) Binary count per sector (min/max via midpoint)
for s, sdata in sector_data.items():
    if "min_bonds" in sdata and "max_bonds" in sdata:
        coeffs = {i: 1.0 for i, b in enumerate(bonds) if b["sector"] == s}
        mid = 0.5*(sdata["min_bonds"] + sdata["max_bonds"])
        H += equality_penalty_ising(coeffs, mid, WEIGHTS["binary_constraint"], num_qubits)

# === Objective: match sector targets ===
for s, sdata in sector_data.items():
    coeffs = {i: b["pmv_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
    H += equality_penalty_ising(coeffs, sdata["target"]/BENCHMARK_VALUE, WEIGHTS["pmv_objective"], num_qubits)

for s, sdata in sector_data.items():
    if "dxs_target" in sdata:
        coeffs = {i: b["dxs_contrib"]/BENCHMARK_VALUE for i, b in enumerate(bonds) if b["sector"] == s}
        H += equality_penalty_ising(coeffs, sdata["dxs_target"]/BENCHMARK_VALUE, WEIGHTS["dxs_objective"], num_qubits)

H = H.simplify()
const_off = constant_shift(H)

print(f"\nHamiltonian built on {num_qubits} qubits.")
print(f"Constant shift (energy offset): {const_off:.6f}")

# ====== VQE setup======
ansatz = RealAmplitudes(num_qubits, entanglement="linear", reps=5)
optimizer = SPSA(maxiter=50)
estimator = Estimator()

energy_hist = []

def callback(eval_count, params, value, std):
    energy_hist.append(float(np.real(value)))
    if eval_count % 10 == 0:
        print(f"Iter {eval_count:3d} | energy {value:.6f}")

initial_point = np.random.rand(ansatz.num_parameters)

vqe = VQE(estimator, ansatz, optimizer, initial_point=initial_point, callback=callback)
res = vqe.compute_minimum_eigenvalue(H)

print("\n=== VQE Result ===")
print(f"Optimal energy (incl. const): {res.eigenvalue.real:.6f}")
print(f"Optimal params length: {len(res.optimal_point)}")

# ====== Plots energy vs iteration ======
loss_hist = [e - const_off for e in energy_hist]

# Energy vs iteration
fig1, ax1 = plt.subplots(figsize=(8, 5))
ax1.plot(range(1, len(energy_hist) + 1), energy_hist, linewidth=2)
ax1.set_xlabel("Iteration")
ax1.set_ylabel("Energy")
ax1.set_title("VQE Energy vs. Iteration")
ax1.grid(True)
fig1.tight_layout()
plt.show()                           
fig1.savefig("energy_vs_iteration.png", dpi=180)

# Loss vs iteration
fig2, ax2 = plt.subplots(figsize=(8, 5))
ax2.plot(range(1, len(loss_hist) + 1), loss_hist, linewidth=2)
ax2.set_xlabel("Iteration")
ax2.set_ylabel("Loss (Energy - constant)")
ax2.set_title("VQE Loss vs. Iteration")
ax2.grid(True)
fig2.tight_layout()
plt.show()                             
fig2.savefig("loss_vs_iteration.png", dpi=180)

print("Saved: energy_vs_iteration.png, loss_vs_iteration.png")

# ====== Sample the optimized state and print top portfolios ======
sampler = Sampler()            
opt_circuit = ansatz.assign_parameters(res.optimal_point)

# Add classical bits + measurements
measured = opt_circuit.copy()
measured.measure_all()

shots = 10000
qres = sampler.run([measured], shots=shots).result()
qd = qres.quasi_dists[0]

num_bits = num_qubits
# Convert to probabilities dict
probs = (qd.binary_probabilities(num_bits=num_bits)
         if hasattr(qd, "binary_probabilities")
         else {(format(k, f"0{num_bits}b") if isinstance(k, int) else k): float(v)
               for k, v in dict(qd).items()})
counts = {b: int(round(p * shots)) for b, p in probs.items()}

def selected_indices_from_bitstring(bits):
    bits_be = bits[::-1] 
    return [i for i, bit in enumerate(bits_be) if bit == "1"]

top = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTop candidate portfolios:")
for rank, (bits, cnt) in enumerate(top, start=1):
    sel = selected_indices_from_bitstring(bits)
    pmv_by_sector = {s: 0.0 for s in sector_data}
    dxs_by_sector = {s: 0.0 for s in sector_data}
    total_pmv = 0.0
    count_by_sector = {s: 0 for s in sector_data}

    for i in sel:
        b = bonds[i]
        s = b["sector"]
        pmv_by_sector[s] += b["pmv_contrib"]
        dxs_by_sector[s] += b["dxs_contrib"]
        count_by_sector[s] += 1
        total_pmv += b["pmv_contrib"]

    print(f"\n#{rank}  freq={cnt/shots:.2%}  selected={len(sel)}  bitstring={bits}")
    print(f"   Total PMV: {total_pmv:,.2f}  (cash in [{cash_min:,.2f}, {cash_max:,.2f}])")

    for s, sd in sector_data.items():
        tgt = sd.get("target", 0.0)
        pmv = pmv_by_sector[s]
        dxs = dxs_by_sector[s]
        cnt_s = count_by_sector[s]
        print(f"   {s}: PMV={pmv:,.2f} (target {tgt:,.2f}) "
              f"| DXS={dxs:,.2f} "
              f"| count={cnt_s}")

# --- Final bond selection
best_bits = max(counts.items(), key=lambda kv: kv[1])[0]
best_sel = selected_indices_from_bitstring(best_bits)

print("\nBond Selection:")
for i, b in enumerate(bonds):
    print(f"  {b['id']}: {1 if i in best_sel else 0}")
```
